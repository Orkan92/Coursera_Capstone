**Now we import our List from Wikipeadia. With the head function we make sure, that the needed data is imported.**

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
df.head(10)
```

**One of our tasks is to replace the "not assigned" cells from the column neighbourhood with the data in the cells of Borough. With the following function we make sure, that all data from Borough is transferred to Neighbourhood, if Neighbourhood contains a "Not assigned".**

```python
df['Neighbourhood'] = np.where(df['Neighbourhood'] == 'Not assigned', df['Borough'], df['Neighbourhood'])
df.head(10)
```

**In the following step we will filter our dataset, for every cell which contains "Not assigned" in Borough and Neighbourhood. Because we matched Borough to Neighbourhood earlier, we make sure that there is not any "Not assigned" left out wrongly. We check that with the head function.**

```python
df_filtered = df[~((df['Borough'] == 'Not assigned') | 
                   (df['Neighbourhood'] == 'Not assigned'))]
```
